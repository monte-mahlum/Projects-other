In this task, we are given a txt file containing a list of words indexed by a number (each line looks like "n word"). The numbers are not in order. To decryp the message, we must stack the words in a pyramid going from smallest number at the top to largest number in the bottom right. The secret message will then be the sum of the last word from each stack in the pyramid. 

e.g., 

Input:
6 bye
52 world
3 hello

Pyramid: 

 hello
 
bye world


Decrypted output:

hello world

```python
def convert_to_array(text):
    data = []
    lines = text.strip().split('\n')  # Split the text into lines
    for line in lines:
        parts = line.split()
        if len(parts) >= 2:
            number = int(parts[0])
            word = parts[1]
            data.append([number, word])
    return data
```

```python
#The same staircase function from before only now also accepts imperfect pyramids, but ignores their imperfect final line
def create_pyramid(array):
    step = 1
    subsets = []
    
    while len(array) != 0:

        if len(array) >= step:
            subsets.append(array[0:step])
            array = array[step:]
            step += 1

        else:
            array = []
            

    return subsets
```

```python
def decrypt(message_file):
    
    #Sorts from smallest to largest index number while reading data
    sorted_data = convert_to_array(message_file)
    sorted_data.sort(key=lambda x: x[0]) 
    
    #Ensure there are no two of the same number
    for i in range(len(sorted_data)-1):
        if sorted_data[i][0] == sorted_data[i+1][0]:
            return "The message is not well-defined"
    
    pyramid = create_pyramid(sorted_data)
    
    message = ""
    for block in pyramid:
        message = message + " " + block[-1][1]
    
    return message
```
